```python
import cv2
# Leer imagen
img = cv2.imread(r"MARS\images\030_D15_day32.png")

# OpenCV lee como BGR, no RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Extraer bandas
R = img[:, :, 0]
G = img[:, :, 1]
B = img[:, :, 2]

# Mostrar
fig, ax = plt.subplots(1,3, figsize=(15,6))

ax[0].imshow(R, cmap='gray')
ax[0].set_title("Rojo")

ax[1].imshow(G, cmap='gray')
ax[1].set_title("Verde")

ax[2].imshow(B, cmap='gray')
ax[2].set_title("Azul")

plt.show()
```

```python
def color_extraction(route):
    """"
    Function to extract the color channels from an image and display them separately.
    """
    img = cv2.imread(route)

    # OpenCV lee como BGR, no RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Extraer bandas
    R = img[:, :, 0]
    G = img[:, :, 1]
    B = img[:, :, 2]
    return R, G, B
```

```python
from pathlib import Path

folder = Path("MARS/images")
# Dictionary to store all bands for each image
data = {}

for file in folder.glob("*.png"):

    route = str(file)
    R, G, B = color_extraction(route)
    data[file.stem] = {"R": R, "G": G, "B": B}
    


```

```python
# To access the bands of a specific image, for example "030_D15_day32":
R_30_D15_day32 = data["030_D15_day32"]["R"]
G_30_D15_day32 = data["030_D15_day32"]["G"]
B_30_D15_day32 = data["030_D15_day32"]["B"]


fig, axs = plt.subplots(3,1, figsize=(10,15))

axs[0].imshow(R_30_D15_day32, cmap='gray')
axs[0].set_title("Rojo")

axs[1].imshow(G_30_D15_day32, cmap='gray')
axs[1].set_title("Verde")

axs[2].imshow(B_30_D15_day32, cmap='gray')
axs[2].set_title("Azul")


plt.show()
```
